```python
# Variabel dan Tipe Data
nama = "No one knows my name"
usia = 24
fokus_ai = True

print(f"Nama: {nama}")
print(f"Usia: {usia}")
print("Fokus AI:", fokus_ai)

# List
misi = ["Meningkatkan kapasitas manusia", "Membangun solusi AI", "Mengatasi dampak destruktif teknologi"]

print("Misi utama:", misi[0] + ", " + misi[1] + ", dan " + misi[2])



# Dictionary
waktu = {
    "subuh": "Refleksi & sholat",
    "pagi": "Belajar intensif",
    "siang": "Eksekusi proyek",
    "sore": "Olahraga",
    "malam": "Review & tidur"
}
print("Rencana pagi:", waktu["pagi"])
print("Rencana siang:", waktu["siang"])
print("Rencana sore:", waktu["sore"])
print("Rencana malam:", waktu["malam"])


```

**🧩 1. Variabel dan Tipe Data Dasar**


1.   nama = "No one knows my name"
2.   usia = 24
3. fokus_ai = True





🔍 Penjelasan:

nama: tipe data string (teks)

usia: tipe data integer (bilangan bulat)

fokus_ai: tipe data boolean (True/False)

Kamu sedang menyimpan state atau kondisi awal dari "entitas" bernama kamu sendiri.




📤 Output:

1.   print(f"Nama: {nama}")
2.   print(f"Usia: {usia}")
3. print("Fokus AI:", fokus_ai)


f"..." disebut f-string (formatted string) → digunakan untuk menyisipkan nilai variabel ke dalam teks.


---

misi = [
  "Meningkatkan kapasitas manusia",

print("Misi utama:", misi[0] + ", " + misi[1] + ", dan " + misi[2])

*🧩 3. Dictionary: Data Bertipe Key–Value*

waktu = {
    "subuh": "Refleksi & sholat",
    "pagi": "Belajar intensif",
    "siang": "Eksekusi proyek",
    "sore": "Olahraga",
    "malam": "Review & tidur"
}

🔍 Penjelasan:

Dictionary menyimpan pasangan data: key (kunci) → value (nilai)

Akses: waktu["pagi"] akan memberikan "Belajar intensif"

📤 Output:
print("Rencana pagi:", waktu["pagi"])
print("Rencana siang:", waktu["siang"])
print("Rencana sore:", waktu["sore"])
print("Rencana malam:", waktu["malam"])


Hasilnya:

Rencana pagi: Belajar intensif
Rencana siang: Eksekusi proyek
Rencana sore: Olahraga
Rencana malam: Review & tidur


Dictionary ini cocok untuk hal-hal yang tidak urut tapi perlu nama jelas (seperti waktu, role, kategori, dsb.)
