```python
# Menginstall library jika belum ada (cukup dijalankan 1x)
!pip install pandas scikit-learn

# Import library yang dibutuhkan
import pandas as pd  # Untuk membaca dan mengolah data
from sklearn.model_selection import train_test_split  # Untuk membagi data ke data latih dan uji
from sklearn.feature_extraction.text import CountVectorizer  # Untuk mengubah teks jadi angka
from sklearn.linear_model import LogisticRegression  # Model AI yang digunakan
from sklearn.metrics import accuracy_score  # Untuk mengukur/cek akurasi model

```

```python
# Membuat data teks dan label (1 = positif, 0 = negatif)
data = {
    'text': [
        'Saya sangat bahagia hari ini',
        'Hari ini buruk sekali',
        'Aku merasa sedih dan hancur',
        'Hidupku luar biasa menyenangkan',
        'Aku benci semuanya',
        'Aku sangat bersyukur atas hari ini',
        'Saya sangat marah dan kecewa',
        'Aku merasa damai dan tenang',
    ],
    'label': [1, 0, 0, 1, 0, 1, 0, 1]
}

# Mengubah data ke bentuk tabel
df = pd.DataFrame(data)

# Menampilkan data
df
```

```python
# Ubah teks ke angka menggunakan metode "Bag of Words"
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])

# Label emosi
y = df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Membuat model
model = LogisticRegression()

# Melatih model dengan data latih
model.fit(X_train, y_train)
```

```python
# Menggunakan model untuk memprediksi data uji
y_pred = model.predict(X_test)

# Menghitung akurasi model
print("Akurasi Model:", accuracy_score(y_test, y_pred))
```

```python
# Kalimat baru untuk dites
input_text = ["Saya kecewa dengan hasilnya", "Saya sangat bangga dan bahagia"]

# Ubah ke angka
input_vec = vectorizer.transform(input_text)

# Prediksi
predictions = model.predict(input_vec)

# Tampilkan hasil
for text, label in zip(input_text, predictions):
    status = "Positif" if label == 1 else "Negatif"
    print(f"Teks: '{text}' → Sentimen: {status}")
```
